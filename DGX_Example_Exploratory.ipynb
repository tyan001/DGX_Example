```python
import tensorflow as tf
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model("my_model/")
```

```python
model.summary()
```

```python
file = open("log.json")
history = json.load(file)
```

```python
y_vloss = history['val_loss']
y_loss = history['loss']

x_len = np.arange(len(y_loss))

plt.figure(figsize=(10, 5))
plt.plot(x_len, y_vloss, marker='^', color='r', label='trainset')
plt.plot(x_len, y_loss, marker='^', color='b', label='trainset')
plt.grid()
plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
```

```python
y_vloss = history['val_accuracy']
y_loss = history['accuracy']

x_len = np.arange(len(y_loss))

plt.figure(figsize=(10, 5))
plt.plot(x_len, y_vloss, marker='^', color='r', label='trainset')
plt.plot(x_len, y_loss, marker='^', color='b', label='trainset')
plt.grid()
plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
```
